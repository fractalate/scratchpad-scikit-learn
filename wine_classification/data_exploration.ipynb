```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split

color_map = {
    'red': 'maroon',
    'white': 'gray',
}
```

```python
df = pd.read_csv('Wine_Quality_Data.csv', header = 0)
df[:10]
```

```python
plt.scatter(
    df['fixed_acidity'], df['volatile_acidity'],
    c = df['color'].map(color_map),
    s=1,
)

plt.title('Acidity Profile by Color')
plt.xlabel('Fixed Acidity')
plt.ylabel('Volatile Acidity')

plt.show()
```

```python
plt.scatter(
    df['residual_sugar'], df['alcohol'],
    c = df['color'].map(color_map),
    s=1,
)

plt.title('Sugar and Alcohol by Color')
plt.xlabel('Residual Sugar')
plt.ylabel('Alcohol')

plt.show()
```
